```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
import json

with open('twitter2022.json') as f:
    data_2022 = json.load(f)

with open('twitter2023.json') as f:
    data_2023 = json.load(f)
```

```python
df_incl_2022=pd.DataFrame(data_2022['includes']['users'])
df_incl_2023=pd.DataFrame(data_2022['includes']['users'])# tienes que hacer un merge con el df_data 

# df_da_metrics_2022=pd.DataFrame(data_2022['data']['public_metrics'])
# df_da_metrics_2023=pd.DataFrame(data_2023['data']['public_metrics'])

df_data_2022=pd.DataFrame(data_2022['data'])
df_data_2023=pd.DataFrame(data_2023['data'])
```

```python
df_data_2022.info()
```

```python
df_data_2022['created_at'] = df_data_2022['created_at'].astype(str)
date=[]

for fecha_str in df_data_2022['created_at']:
    # Utiliza strptime para analizar la cadena en formato de fecha y hora
    fecha_datetime = datetime.strptime(fecha_str, '%Y-%m-%dT%H:%M:%S.%fZ')
    date.append(fecha_datetime)
df_data_2022['date_time']=date
df_data_2022.head()
```

```python
df_data_2023['created_at'] = df_data_2023['created_at'].astype(str)
date=[]

for fecha_str in df_data_2023['created_at']:
    # Utiliza strptime para analizar la cadena en formato de fecha y hora
    fecha_datetime = datetime.strptime(fecha_str, '%Y-%m-%dT%H:%M:%S.%fZ')
    date.append(fecha_datetime)
df_data_2023['date_time']=date
df_data_2023.head()
```

```python
df_data_2022['public_metrics'] = df_data_2022['public_metrics'].astype(str)

df_data_2022['impression_count']=df_data_2022['public_metrics'].str.split(',').str.get(0)
df_data_2022['impression_count']=df_data_2022['impression_count'].str.split(':').str.get(1)

df_data_2022['like_count']=df_data_2022['public_metrics'].str.split(',').str.get(1)
df_data_2022['like_count']=df_data_2022['like_count'].str.split(':').str.get(1)

df_data_2022['quote_count']=df_data_2022['public_metrics'].str.split(',').str.get(2)
df_data_2022['quote_count']=df_data_2022['quote_count'].str.split(':').str.get(1)

df_data_2022['reply_count']=df_data_2022['public_metrics'].str.split(',').str.get(3)
df_data_2022['reply_count']=df_data_2022['reply_count'].str.split(':').str.get(1)

df_data_2022['retweet_count']=df_data_2022['public_metrics'].str.split(',').str.get(4)
df_data_2022['retweet_count']=df_data_2022['retweet_count'].str.split(':').str.get(1)
df_data_2022['retweet_count'].replace(regex='}', value='',inplace=True)

df_data_2022[['impression_count','like_count','quote_count','reply_count','retweet_count']] = df_data_2022[['impression_count','like_count','quote_count','reply_count','retweet_count']].astype(int)


df_data_2022.head()
```

```python
df_data_2023['public_metrics'] = df_data_2023['public_metrics'].astype(str)

df_data_2023['impression_count']=df_data_2023['public_metrics'].str.split(',').str.get(0)
df_data_2023['impression_count']=df_data_2023['impression_count'].str.split(':').str.get(1)

df_data_2023['like_count']=df_data_2023['public_metrics'].str.split(',').str.get(1)
df_data_2023['like_count']=df_data_2023['like_count'].str.split(':').str.get(1)

df_data_2023['quote_count']=df_data_2023['public_metrics'].str.split(',').str.get(2)
df_data_2023['quote_count']=df_data_2023['quote_count'].str.split(':').str.get(1)

df_data_2023['reply_count']=df_data_2023['public_metrics'].str.split(',').str.get(3)
df_data_2023['reply_count']=df_data_2023['reply_count'].str.split(':').str.get(1)

df_data_2023['retweet_count']=df_data_2023['public_metrics'].str.split(',').str.get(4)
df_data_2023['retweet_count']=df_data_2023['retweet_count'].str.split(':').str.get(1)
df_data_2023['retweet_count'].replace(regex='}', value='',inplace=True)

df_data_2023[['impression_count','like_count','quote_count','reply_count','retweet_count']] = df_data_2023[['impression_count','like_count','quote_count','reply_count','retweet_count']].astype(int)


df_data_2023.head()
```

```python
df_data_2022=df_data_2022.sort_values('date_time', ascending=True)
df_data_2023=df_data_2023.sort_values('date_time', ascending=True)
df_data_tweets=pd.concat([df_data_2022,df_data_2023],ignore_index=True)

df_data_tweets=df_data_tweets.drop(['created_at','public_metrics'],axis=1)
```

```python
df_data_tweets.to_csv('tweets_TheBridge.csv')
```
